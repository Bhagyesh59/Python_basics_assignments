```python
3. Compare and contrast map, filter, and reduce.

- map: map is used to apply a given function to each element in an iterable and create a new iterable containing the results.
       It's perfect for transforming data by performing the same operation on every item in the input sequence.

filter: filter is designed for selecting elements from an iterable based on a specified condition, defined by a predicate 
        function. It returns a new iterable containing only the elements that meet the condition, allowing you to filter 
        data based on a specific criterion.

reduce: reduce is used to cumulatively apply a binary function to the items of an iterable, aggregating the data into a 
        single result. It's ideal for tasks like calculating the sum or product of a list of values by repeatedly applying 
        the function to pairs of elements.
```

```python
4. What are function annotations, and how are they used?

- Function annotations are a feature in Python that allows you to attach metadata or additional information to the 
parameters and return value of a function. These annotations are optional and do not affect the function's behavior 
at runtime. They are primarily used for documenting the expected types of function arguments and return values, making 
code more self-explanatory and aiding in type checking and code analysis.
Function annotations are defined using colons (:) and follow the parameter name for input arguments and the -> symbol
for the return value. Here's an example of a function with annotations:


def add(a: int, b: int) -> int:
    """
    This function takes two integers, 'a' and 'b', and returns their sum.
    """
    return a + b
```

```python
5. What are recursive functions, and how are they used?

- A recursive function is a function that calls itself during its execution. 
  This means that the function will continue to call itself and repeat its behavior until some condition is met to return a result


# example
def fact(x):
    if x == 1 :
        return 1
    else :
        return x * fact(x-1) # recurtion
    
fact(3)
```

```python
 6. What are some general design guidelines for coding functions?
    
- Clear Purpose: Ensure each function has a clear and specific purpose, performing one task or operation.

  Descriptive Names: Use meaningful and descriptive names for functions that convey their intent.

  Modularity: Design functions to be modular and reusable by encapsulating logic into smaller functions.

  Function Length: Keep functions relatively short and focused to improve readability.

  Parameter Limit: Limit the number of parameters a function accepts for clarity and ease of use.

  Documentation: Provide clear docstrings to explain the function's purpose, parameters, and return values.

  Error Handling: Handle exceptions gracefully within functions and raise specific exceptions with helpful error messages when needed.

  Return Values: Ensure functions return values that match their intended purpose; be consistent in what is returned.

  Side Effects: Minimize side effects (e.g., global variable changes) within functions for predictability.

  Testing: Write unit tests to validate function behavior and catch errors early during development.
 




```

```python
7. Name three or more ways that functions can communicate results to a caller.

- Return Values: Functions can use the return statement to send a value or a collection of values back to the caller. 
                 The caller can capture these return values and use them for further computation or processing.

def add(a, b):
    return a + b
result = add(2, 3)


Global Variables: Functions can modify global variables to communicate results. While this approach is generally
                  discouraged due to potential side effects, it's a way to share information between functions and the global scope.
        
total = 0
def add_to_total(value):
    global total
    total += value
add_to_total(5)  



Print Statements: Functions can use print statements to display results or intermediate information. While not suitable 
                  for returning values for further computation, it's a way to provide visual feedback.
        
def greet(name):
    print(f"Hello, {name}!")
greet("Boss") 



Mutable Objects: Functions can modify mutable objects (e.g., lists, dictionaries) passed as arguments, allowing them 
                 to communicate results by changing the state of these objects.

def append_element(lst, element):
    lst.append(element)
my_list = [1, 2, 3]
append_element(my_list, 4)



Exceptions: Functions can communicate exceptional conditions or errors to the caller by raising exceptions using the 
            raise statement. The caller can then handle the exception as needed.

def divide(a, b):
    if b == 0:
        raise ValueError("Division by zero is not allowed obviously")
    return a / b
try:
    result = divide(10, 0)  
except ValueError as e:
    print(f"Error: {e}")
```
